```python
import random
import pygame

# Set up the game board
board_width = 20
board_height = 20
cell_size = 20
board = [[0 for x in range(board_width)] for y in range(board_height)]

# Initialize the snake
snake = [(board_width // 2, board_height // 2)]
snake_direction = (0, -1)

# Generate food
def generate_food():
    x = random.randint(0, board_width - 1)
    y = random.randint(0, board_height - 1)
    while board[y][x] != 0:
        x = random.randint(0, board_width - 1)
        y = random.randint(0, board_height - 1)
    board[y][x] = 2
generate_food()

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((board_width * cell_size, board_height * cell_size))
clock = pygame.time.Clock()

# Game loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and snake_direction != (0, 1):
                snake_direction = (0, -1)
            elif event.key == pygame.K_DOWN and snake_direction != (0, -1):
                snake_direction = (0, 1)
            elif event.key == pygame.K_LEFT and snake_direction != (1, 0):
                snake_direction = (-1, 0)
            elif event.key == pygame.K_RIGHT and snake_direction != (-1, 0):
                snake_direction = (1, 0)
    
    # Move the snake
    head_x, head_y = snake[-1]
    new_head = (head_x + snake_direction[0], head_y + snake_direction[1])
    if new_head[0] < 0 or new_head[0] >= board_width or new_head[1] < 0 or new_head[1] >= board_height:
        running = False
    elif board[new_head[1]][new_head[0]] == 1:
        running = False
    elif board[new_head[1]][new_head[0]] == 2:
        snake.append(new_head)
        board[new_head[1]][new_head[0]] = 1
        generate_food()
    else:
        snake.append(new_head)
        board[new_head[1]][new_head[0]] = 1
        tail_x, tail_y = snake.pop(0)
        board[tail_y][tail_x] = 0
    
    # Render the game board
    screen.fill((0, 0, 0))
    for y in range(board_height):
        for x in range(board_width):
            if board[y][x] == 1:
                pygame.draw.rect(screen, (255, 255, 255), (x * cell_size, y * cell_size, cell_size, cell_size))
            elif board[y][x] == 2:
                pygame.draw.rect(screen, (255, 0, 0), (x * cell_size, y * cell_size, cell_size, cell_size))
    pygame.display.flip()
    
    # Delay for consistent frame rate
    clock.tick(10)

# Clean up Pygame
pygame.quit()
```
